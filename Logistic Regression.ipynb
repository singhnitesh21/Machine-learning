# Logistic Regression

## Hypothesis

$ H_{\theta}(X) = \frac{1}{1 + e^{-\theta^{T}X}}  = \frac{1}{1 + e^{-(\theta_{0}X_{0}^{i} + \theta_{1}X_{1}^{i} + ... + \theta_{n}X_{n}^{i})}}$

```python
def sigmoid(Value):
    return (1.0/ (1.0 + np.exp(-Value)))
```

```python
def Hypothesis(X, theta):
    return sigmoid(np.dot(X, theta))
```

```python
example = np.linspace(-10, 10, 20)
plt.plot(example, sigmoid(example), label = "Plot of Sigmoid")
plt.legend();
```

## Error

$ Error = -\frac{1}{m}\sum_{i=1}^{m}(Y^{i}log(h^{i}) + ((1 - Y^{i})log(1 - h^{i})))$

```python
def error(X, Y, theta):
    hx = Hypothesis(X, theta)
    
    error = -1 * np.mean((Y * np.log(hx)) + ((1 -Y) * np.log(1 - hx)))
    return error
    
```

## Descent
$ Descent = \sum_{i=1}^{m}(Y^{i} - h_{\theta}(X^{i}))X^{i}_{j} $

```python
def Descent(X, Y, theta):
    m = X.shape[0]
    hx = Hypothesis(X, theta)
    grad = np.dot(X.T, (Y - hx))
    
    return grad/m
    
```

# Gradient Descent

$ Gradient Descent :
\theta = \theta +  \alpha\sum_{i=1}^{m}(Y^{i} - h_{\theta}(X^{i}))X^{i}_{j} = \theta + \alpha Gradient$

```python
def GradientDescent(X, Y, lr = 0.1, max_itr = 500):
    
    n = X.shape[1]
    error_list = []
    theta = np.zeros((n, 1))
    
    for i in range(max_itr):
        error_list.append(error(X, Y, theta))
        
        grad = Descent(X, Y, theta)
        theta = theta + lr * grad
    
    return theta, error_list
            
```

```python
# Adding Bias Column
ones = np.ones((X_train.shape[0], 1))

X_new_train = np.hstack((ones, X_train))
print(X_new_train.shape)
```

$\theta_{0} + \theta_{1}X_{1} + \theta_{2}X_{2} = 0$

$\theta_{2}X_{2} = - \theta_{0} - \theta_{1}X_{1} $

$X_{2} = \frac{- \theta_{0} - \theta_{1}X_{1}}{\theta_{2}} $

# Prediction over Testing Dataset

```python
def predict(X, theta):
    
    h = Hypothesis(X, theta)
    output = np.zeros((X.shape[0], 1))
    
    output[h >= 0.5] = 1 # placing 1 where hypothesis >= 0.5
    output = output.astype("int_")
    return output
```

```python
predict_train = predict(X_new_train, theta)
```

```python
# Predicting over test data

ones = np.zeros((X_test.shape[0], 1))
X_new_test = np.hstack((ones, X_test))

predict_test = predict(X_new_test, theta)
```

## Accuracy

```python
def accuracy(actual, predict):
    actual = actual.astype('int_')
    
    acc = (np.sum(actual == predict)) / actual.shape[0]
    acc *= 100
    return acc
```

```python
# Training Accuracy
accuracy(Y_train, predict_train)
```

```python
# Testing Accuracy
Y_test =  Y_test.reshape((-1, 1))
accuracy(Y_test, predict_test)
```
